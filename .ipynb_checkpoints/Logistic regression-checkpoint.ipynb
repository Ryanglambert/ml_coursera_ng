# Motivation for different cost function for Log Reg

```octave
function logit = logitFunction (z)
    logit = 1.0 ./ (1 + exp(-z));
endfunction
```

```octave
x = linspace(-10, 10, 100);
plot(x, logitFunction(x))
```

##### Cost function for linear regression (Mean Squared Error)

```octave
function cost = SSE(theta, X, y)
    cost = sum((theta' * X - y) .^ 2)
endfunction
```

### How does the Cost Function (SSE) look for linear regression

```octave
actual_thetas = [1, 1]';
X = rand(100, 1);
X_w_const = [ones(100, 1), X + .1 .* randn(length(X),1)];
y =  X_w_const * actual_thetas;
scatter(X, y);
```

```octave
theta_space_0 = linspace(actual_thetas(1) - 3, actual_thetas(1) + 3, 40)';
theta_space_1 = linspace(actual_thetas(2) - 3, actual_thetas(2) + 3, 40)';
[theta_grid_0, theta_grid_1] = meshgrid(theta_space_0, theta_space_1);
theta_grid = [reshape(theta_grid_0, [], 1), reshape(theta_grid_1, [], 1)];
costs = sum((X_w_const * theta_grid' - y) .^ 2, 1);
mesh(theta_grid_1, theta_grid_0, reshape(costs, 40, 40));
```

Looks convex!

### How does the Cost Function (SSE) look for logistic regression?

```octave
X = linspace(-2, 2, 10)';
y = [0, 0, 0, 0, 1, 1, 0, 1, 1, 1]';
thetas = [0, 1]'
X_w_const = [ones(10, 1), X];
plot(X, 1/1+(e.^(-thetas*X_w_const)));
scatter(X, y);
```

not what we're expecting!

```octave
spread = 10
theta_space_0 = linspace(thetas(1) - spread, thetas(1) + spread, 40)';
theta_space_1 = linspace(thetas(2) - spread, thetas(2) + spread, 40)';
[theta_grid_0, theta_grid_1] = meshgrid(theta_space_0, theta_space_1);
theta_grid = [reshape(theta_grid_0, [], 1), reshape(theta_grid_1, [], 1)];
costs = sum((X_w_const * theta_grid' - y) .^ 2, 1);
mesh(theta_grid_1, theta_grid_0, reshape(costs, 40, 40));
```

Can't seem to make this non-convex. Perhaps this happens more readily at higher dimensions

```octave
function cost = logitCost(theta, X, y)
    # cost if y=1 -log(H_theta(x))
    h = theta'*X
    cost = y .* -log(h) + (1 - y) * -log(1 - h) 
endfunction
```
